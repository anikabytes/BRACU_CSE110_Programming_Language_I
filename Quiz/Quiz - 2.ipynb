## Question-1
Write a python program that takes a string as an input from the user. If 
the sum of the first half of the input is equal to the sum of the last half 
of the input, then it prints â€œYou won the lottery!â€, otherwise it prints 
â€œYou lost.â€


The input can be digits only (0-9). If the length of the string is even, 
then you need to consider all numbers. If the length is odd, you can ignore 
the middle number.

**<font color='red'>[You are not allowed to use the built-in sum() and slice() for this task]</font>**

===================================================================

**Sample Input 1:**
37938725

**Sample Output 1:**
You won the lottery!

**Explanation 01:**
The first half of the input = 3+7+9+3 = 22.
The last half of the input = 8+7+2+5 = 22.
The summation is equal. Since the length of the string is even, we 

considered all numbers.

===================================================================

**Sample Input 2:**
7357314

**Sample Output 2:**
You lost.

**Explanation 02:**
The first half of the input = 7+3+5 = 15
The last half of the input = 3+1+4 = 8
The summation is not equal. Since the length of the string is odd, we 
ignored the middle number, 5.

```python
#todo
lottery=input('enter the lottery number:')
summ1=0
summ2=0
if(len(lottery)%2==0):
    for number in range(0,len(lottery)//2):
        summ1+=int(lottery[number])
    for number in range(len(lottery)//2,len(lottery)):
        summ2+=int(lottery[number])
else:
    for number in range(0,(len(lottery)//2)):
        summ1+=int(lottery[number])
    for number in range((len(lottery)//2)+1,len(lottery)):
        summ2+=int(lottery[number])
if(summ1==summ2):
    print('You have won')
else:
    print('You lost')
```

Write a python program that will take a string as an input from the user. 
Your task is to check the ASCII value of every character. If the ASCII 
value of the character is even, replace that character with the previous 
index character of the original string; or else keep the same.
For the 0th index, since there is no previous character replace it with â€˜1â€™

**<font color='red'>[You cannot use the built-in replace() function.]</font>**

================================

**Sample Input 1:**
Python

**Sample Output 1:**
1yytoo

**Explanation 01:**
The ASCII value of 'P'=80, 'y'=121, 't'=116, 'h'=104, 'o'=111, 'n'=110. 
Since the ASCII of â€˜Pâ€™ is even, it is replaced by â€˜1â€™. In the same way, the 
ASCII of â€˜tâ€™ is even, so replaced by the previous character â€˜yâ€™. Next, the 
ASCII of â€˜hâ€™ is even, so replaced by the previous character â€˜tâ€™. Finally, 
the ASCII of â€˜nâ€™ is even, so replaced by the previous character â€˜oâ€™.

================================

**Sample Input 2:**
grow

**Sample Output 2:**
ggow

**Explanation 02:**
The ASCII value of 'g'=103, 'r'=114, 'o'=111, 'w'=119. Since the ASCII of 
â€˜râ€™ is even, it was replaced by the previous character â€˜gâ€™. The other 
character has an odd ASCII value, so they were the same


```python
#todo
word=input('enter a word:')
ascii_word=''
new_word=''
for char in range(0,len(word)):
    ascii_word=ord(word[char])
    if(ascii_word%2==0):
        if(char==0):
            new_word+='1'
        else:
            new_word+=chr(ord(word[char-1]))
    if(ascii_word%2!=0):
        new_word+=word[char]
print(new_word)
```
